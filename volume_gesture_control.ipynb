```python
wCam , hCam = 720,640

cap = cv2.VideoCapture(0)
cap.set(3,wCam)
cap.set(4,hCam)
# pTime = 0
```

```python
detector = htm.handDetector(detectionCon=0.7)
```

```python
p_time=0
while True:
    success,img = cap.read()
    if not success:
        break
    img,results=detector.findHands(img)
    landmark_list = detector.findPosition(img,draw=False)
    if len(landmark_list) != 0:
        # print(landmark_list[8],landmark_list[4])
        x1 ,y1 = landmark_list[4][1],landmark_list[4][2]
        x2,y2 = landmark_list[8][1],landmark_list[8][2]
        cv2.circle(img, (x1,y1), 9 ,(255,0,255), cv2.FILLED)
        cv2.circle(img, (x2,y2), 9 ,(255,0,255), cv2.FILLED)
        cv2.line(img,(x1,y1),(x2,y2),(255,0,255),3)
        length = math.hypot(x2-x1,y2-y1)

        #Hand range = 50-300
        # volume range -65 to 0
        #########for windows
        # vol = np.interp(length,[50,300],[minVol,maxVol])
        # print(vol)
        # volume.SetMasterVolumeLevel(vol,None)


        #for ubuntu
        vol = np.interp(length,[50,250],[0,100])
        print(vol)
        command = ["amixer", "set", "Master", f"{vol}%"]
        # Execute the command
        subprocess.run(command)
    cv2.imshow("Image", img)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
